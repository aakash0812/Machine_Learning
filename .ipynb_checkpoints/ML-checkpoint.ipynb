# General Terms/Concepts used in ML

- Supervised vs unsupervised machine learning. The former uses labeled data i.e. the model is trained on set of data that has answers. 

- Supervised Learning:
    - Classification & Regression categories
    - A row in a dataset is called a sample
    - A column in a dataset is called a feature of the sample
    - Each sample has "target" which is the assigned label
    - Classification is more binary - i.e. you are trying to classify a dataset into one or the other. Example: whether an image is a cat or a dog. There could be multi-classification models as well. 
        - Example of classification algorithm is k-nearest neighbor 
    - Regression works with continous data
        - Example of regression algorith is multiple linear regression
- Unsupervised Learning:
     - For unsupervsed one of the algorithms could be clustering by reducing the dimensions i.e. reduce the number of features 

# Using ML on digits, iris and diabetes dataset from Scikit

```python
### FINDING the best k-Neighbors - this is a classification algorith in supervised learning. #########

#Loading the data
from sklearn.datasets import load_iris
iris = load_iris()

#Loading the K-fold, and cross_val
from sklearn.model_selection import KFold, cross_val_score

#Estimation
from sklearn.neighbors import KNeighborsClassifier

for k in range(1, 15):
    kfold = KFold(n_splits=10, random_state=11, shuffle=True)
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(estimator=knn, X=iris.data, y=iris.target, cv=kfold)
    print(f'k={k:<2}; mean accuracy={scores.mean():.2%}; ' + f'standard deviation={scores.std():.2%}')
    
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
#Below is the regression algorithm for SUPERVISED learning ######

from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

import pandas as pd
pd.set_option('display.width', None)

#View the data

diabetes_df = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
diabetes_df['Disease_Progression'] = pd.Series(diabetes.target)

diabetes_df.head


#Visualizing the features
sample_df = diabetes_df.sample(frac=0.1, random_state=17)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=2)
sns.set_style('whitegrid') 
    
for feature in diabetes.feature_names:
    plt.figure(figsize=(16, 9))
    sns.scatterplot(data=sample_df, x=feature, 
    y='Disease_Progression', hue='Disease_Progression', 
    palette='cool', legend=False)


#Splitting the data to train and test for regression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, random_state=11)

#Train the regression model
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X=X_train, y=y_train)

for i, name in enumerate(diabetes.feature_names):
    print(f'{name:>10}: {linear_regression.coef_[i]}')
    
#Test the Model
predicted = linear_regression.predict(X_test)
expected = y_test
predicted[:5]
expected[:5]


#Display the predicted vs expected results of regression
df = pd.DataFrame()
df['Expected'] = pd.Series(expected)
df['Predicted'] = pd.Series(predicted)
figure = plt.figure(figsize=(9, 9))
axes = sns.scatterplot(data=df, x='Expected', y='Predicted', hue='Predicted', palette='cool', legend=False)
start = min(expected.min(), predicted.min())
end = max(expected.max(), predicted.max())
axes.set_xlim(start, end)
axes.set_ylim(start, end)
line = plt.plot([start, end], [start, end], 'k--')


#Evaluating the regression model

from sklearn import metrics
metrics.r2_score(expected, predicted)
metrics.mean_squared_error(expected, predicted)



#Assessing other models
from sklearn.linear_model import ElasticNet, Lasso, Ridge
estimators = {'LinearRegression': linear_regression, 'ElasticNet': ElasticNet(), 'Lasso': Lasso(), 'Ridge': Ridge()}
from sklearn.model_selection import KFold, cross_val_score
for estimator_name, estimator_object in estimators.items():
    kfold = KFold(n_splits=10, random_state=11, shuffle=True)
    scores = cross_val_score(estimator=estimator_object, 
    X=diabetes.data, y=diabetes.target, cv=kfold,
    scoring='r2')
    print(f'{estimator_name:>16}: ' + 
    f'mean of r2 scores={scores.mean():.3f}')
```

```python
### COMPARING PCA versus TSNE dimensionality reduction techniques on digit data set. TSNE is already performed in the book called: "Intro to Python"  ###

#load digits dataset
from sklearn.datasets import load_digits
digits = load_digits()

#Import PCA decomposition
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=11)

#Transforming the dataset into two dimensions and fitting them 
pca.fit(digits.data)
digits_pca = pca.transform(digits.data)

#Visualizing the dataset
import matplotlib.pyplot as plt
dots = plt.scatter(digits_pca[:, 0], digits_pca[:, 1], c=digits.target, cmap=plt.cm.get_cmap('nipy_spectral_r', 10))
colorbar = plt.colorbar(dots)
```

Observation: Both TSNE and PCA give 10 clusters in total. However, with PCA the clusters aren't seggregated as it is with TSNE. On top of that, it is very hard to visualize which digit has the highest frequency. Though PCA provides a better illustration of variance in the digits than TSNE does.  

```python
### COMPARING PCA versus TSNE dimensionality reduction techniques on IRIS data set. PCA is already performed in the book called: "Intro to Python"  ###


#Loading the data
from sklearn.datasets import load_iris
iris = load_iris()

#Import TSNE estimator and reduce it to two dimensions
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=11)
reduced_iris = tsne.fit_transform(iris.data)

#Visualizing the dataset
import numpy as np

unique_labels = np.unique(iris.target)
for label in unique_labels:
    plt.scatter(reduced_iris[iris.target == label, 0], reduced_iris[iris.target == label, 1], label=iris.target_names[label])

plt.legend()
plt.show()
```

Observation: Both graphs show that setosa is completely different from the other two Iris flower species. On top of that, versicolor and virginica are hard to differentiate unlike setosa and both of the are clusters together. With the PCA image, however, it tells that there is a variability in both component2.
